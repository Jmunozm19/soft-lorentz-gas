```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Soft Lorentz Gas (Fermi potentials) - Figure 1 style plots
# Generates:
#   - soft_lorentz_unitcell.png  (potential field in triangular lattice unit cell)
#   - soft_lorentz_profile.png   (1D profile along base with E*, r1(E), W(E))
# Parameters follow the paper's typical choices:
#   r0 = 1.0, sigma = 0.01, w = 0.05, L = 2*r0 + w
# ============================================================

# ---------------- Parameters ----------------
r0 = 1.0
sigma = 0.01
w = 0.05
L = 2*r0 + w                                # lattice spacing
E_example = 0.60                            # Energy for annotations in profile panel (0<E<1)

# Triangular lattice basis vectors
a1 = np.array([L, 0.0])
a2 = np.array([L/2.0, np.sqrt(3)/2.0 * L])

# ---------------- Potentials ----------------
def V_fermi(r, r0=r0, sigma=sigma):
    """Fermi potential of a single circular scatterer."""
    return 1.0 / (1.0 + np.exp((r - r0) / sigma))

def lattice_points(n_range=2):
    """Generate lattice points R = i*a1 + j*a2 in a window [-n, n]x[-n, n]."""
    pts = []
    for i in range(-n_range, n_range+1):
        for j in range(-n_range, n_range+1):
            pts.append(i*a1 + j*a2)
    return np.array(pts)

LPTS = lattice_points(2)

def V_total_xy(x, y):
    """Total potential at (x,y) from scatterers in LPTS."""
    p = np.array([x, y])
    s = 0.0
    for R in LPTS:
        r = np.linalg.norm(p - R)
        s += V_fermi(r)
    return s

# ---------------- Helper formulas (profile) ----------------
def V1_1D(x):
    """1D potential from the scatterer centered at x=0 along the base (y=0)."""
    return 1.0 / (1.0 + np.exp((np.abs(x) - r0)/sigma))

def V2_1D(x):
    """1D potential from the scatterer centered at x=L along the base (y=0)."""
    return 1.0 / (1.0 + np.exp((np.abs(x - L) - r0)/sigma))

def r1_of_E(E):
    """Approximate radius from single-scatterer contribution: r1(E) = sigma*ln(1/E - 1) + r0."""
    E = np.clip(E, 1e-12, 1.0-1e-12)
    return sigma * np.log(1.0/E - 1.0) + r0

# ---------------- Compute profile quantities ----------------
# Energy threshold E* at the mid-gap x = r0 + w/2 using two nearest scatterers
E_star = V1_1D(r0 + w/2) + V2_1D(r0 + w/2)

r1_E = r1_of_E(E_example)          # r1(E) for annotation
W_E  = L - 2.0*r1_E                # overestimates true gap (ignores overlap), but matches paper's r1-usage

# ---------------- Figure: 1D profile ----------------
x = np.linspace(-0.2*L, 1.2*L, 2000)
V1 = V1_1D(x)
V2 = V2_1D(x)
Vsum = V1 + V2

fig2, ax2 = plt.subplots(figsize=(7, 4), dpi=160)
ax2.plot(x/L, V1, label=r"$V_1(x)$", linewidth=1.8)
ax2.plot(x/L, V2, label=r"$V_2(x)$", linewidth=1.8)
ax2.plot(x/L, Vsum, label=r"$V_1(x)+V_2(x)$", linewidth=2.2)
ax2.set_xlim(0.2, 0.8)

# Lines for E* and E_example
ax2.axhline(E_star, color='k', linestyle='--', linewidth=1.2, label=r"$E^\ast$")
ax2.axhline(E_example, color='gray', linestyle=':', linewidth=1.2, label=rf"$E={E_example:.2f}$")

# Mark r1(E) and W(E) on [0, L]
x_r1 = r1_E
x_W_end = r1_E + W_E
ax2.hlines(E_example, x_r1/L, x_W_end/L, colors='gray', linewidth=3.0)
ax2.vlines([x_r1/L, x_W_end/L], ymin=0, ymax=E_example, colors='gray', linestyles=':')
ax2.text((x_r1/L + x_W_end/L)/2, E_example + 0.02, r"$W(E)$", ha='center', va='bottom', fontsize=10)
ax2.text(x_r1/L, 0.02, r"$r_1(E)$", ha='center', va='bottom', fontsize=10, rotation=90)

ax2.set_xlabel(r"$x/L$")
ax2.set_ylabel(r"Potential $V$")
ax2.set_title("Perfil 1D del potencial en la base de la celda")
ax2.set_xlim(0.3, 0.7)
ax2.set_ylim(0, min(2.5, Vsum.max()*1.05))
ax2.legend(loc="upper right", ncol=2, fontsize=9, frameon=False)
fig2.tight_layout()
fig2.savefig("soft_lorentz_profile.png", bbox_inches="tight")

# ---------------- Figure: Potential field in unit cell ----------------
# Parametric grid over u,v in [0,1], p(u,v) = u*a1 + v*a2
Nu, Nv = 220, 220
u = np.linspace(0, 1, Nu)
v = np.linspace(0, 1, Nv)
U, Vv = np.meshgrid(u, v, indexing='xy')
X = U*a1[0] + Vv*a2[0]
Y = U*a1[1] + Vv*a2[1]

# Total potential on grid
Vgrid = np.zeros_like(X)
for R in LPTS:
    dx = X - R[0]
    dy = Y - R[1]
    rr = np.sqrt(dx*dx + dy*dy)
    Vgrid += 1.0 / (1.0 + np.exp((rr - r0) / sigma))

# Boundary of unit cell (parallelogram)
origin = np.array([0.0, 0.0])
corners = np.array([origin, origin + a1, origin + a1 + a2, origin + a2, origin])

fig1, ax1 = plt.subplots(figsize=(6.5, 6), dpi=160)
vmax = min(2.0, np.percentile(Vgrid, 99.8))
cnt = ax1.contourf(X, Y, Vgrid, levels=40, vmin=0.0, vmax=vmax)
ax1.plot(corners[:,0], corners[:,1], color='k', linewidth=1.5)

# Lattice vectors (arrows)
ax1.arrow(0, 0, a1[0], a1[1], head_width=0.05*L, head_length=0.08*L, length_includes_head=True, color='k')
ax1.text(a1[0]*0.6, a1[1]*0.6, r"$\mathbf{a}_1$", fontsize=10)
ax1.arrow(0, 0, a2[0], a2[1], head_width=0.05*L, head_length=0.08*L, length_includes_head=True, color='k')
ax1.text(a2[0]*0.6, a2[1]*0.6, r"$\mathbf{a}_2$", fontsize=10)

# Centers of three nearest scatterers
centers = np.array([[0,0], a1, a2])
ax1.scatter(centers[:,0], centers[:,1], s=30, c='w', edgecolors='k', zorder=3)

# Overlay an approximate equipotential circle at the origin for r1(E_example)
theta = np.linspace(0, 2*np.pi, 400)
xc = r1_E*np.cos(theta)
yc = r1_E*np.sin(theta)
ax1.plot(xc, yc, color='w', linewidth=1.4, alpha=0.9, label=rf"Equipotencial aprox. $r_1(E={E_example:.2f})$")

ax1.set_aspect('equal', 'box')
ax1.set_xlim(-0.2*L, a1[0] + a2[0] + 0.2*L)
ax1.set_ylim(-0.2*L, a1[1] + a2[1] + 0.2*L)
ax1.set_title("Campo de potencial en la celda unitaria (red triangular)")
cbar = fig1.colorbar(cnt, ax=ax1, fraction=0.046, pad=0.04)
cbar.set_label("V")
ax1.legend(loc="upper right", frameon=False, fontsize=9)
fig1.tight_layout()
fig1.savefig("soft_lorentz_unitcell.png", bbox_inches="tight")

# Print some reference values
print("L =", L)
print("E* =", E_star)
print("r1(E_example) =", r1_E)
print("W(E_example)  =", W_E)
```
